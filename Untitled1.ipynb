# 💰 Loan Default Risk Analysis
### Data Analysis Project using Python

This project analyzes loan customers to understand risk factors behind loan default using statistics and visualization.

# 📌 Problem Statement

The goal of this project is to identify key factors that influence loan default and build a risk scoring model to classify customers into Low, Medium, and High risk categories.

# 📊 Dataset Description

The dataset contains customer financial and personal information such as income, credit score, loan amount, and employment details.

Target variable:
- Default (0 = No default, 1 = Default)

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\pc\Documents\projects\analyst dashbords\project1 excel\Loan_default.csv")
```

```python
df.columns = df.columns.str.strip()
df['Default'] = pd.to_numeric(df['Default'], errors='coerce')
df = df.dropna()
```

# 🚨 Risk Factor Analysis

We compare safe and risky customers to identify key differences.

```python

fig, axes = plt.subplots(1, 3, figsize=(18,5))

# Income
sns.histplot(df['Income'], kde=True, ax=axes[0])
axes[0].set_title("Income Distribution")

# Credit Score
sns.histplot(df['CreditScore'], kde=True, ax=axes[1])
axes[1].set_title("Credit Score Distribution")

# Loan Amount
sns.histplot(df['LoanAmount'], kde=True, ax=axes[2])
axes[2].set_title("Loan Amount Distribution")

plt.tight_layout()
plt.show()
```

#  Insights

Uniform Distributions: All three variables—Income, Credit Score, and Loan Amount—show a perfectly flat, uniform distribution.

Credit Scores: The scores cover the standard 300 to 850 range. People with terrible credit scores appear just as often as people with perfect credit scores


```python
safe = df[df['Default'] == 0]
risky = df[df['Default'] == 1]

safe['Income'].mean(), risky['Income'].mean()
```

#  Insights
The dataset is divided into two groups:
Safe (0): Customers who repaid loans
Risky (1): Customers who defaulted

Average Income (Safe group): ~83,900
Average Income (Risky group): ~71,845

Income Difference: ~12,055
Key Insight:
Higher income customers are less likely to default, while lower income customers have a higher risk of loan default.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.hist(safe['LoanAmount'], bins=30, alpha=0.6, label='Safe', color='green')
plt.hist(risky['LoanAmount'], bins=30, alpha=0.6, label='Risky', color='red')

plt.title("Loan Amount: Safe vs Risky Customers")
plt.xlabel("Loan Amount")
plt.ylabel("Count")
plt.legend()
plt.show()
```

#  Insights

The chart compares two groups:
Safe (green): customers who repaid loans
Risky (red): customers who defaulted

Safe customers take loans across all amounts in a balanced way

Risky customers increase as loan amount increases

Key insight: Higher loan amounts have higher default risk, while smaller loans are safer.

```python
plt.figure(figsize=(10,5))

plt.hist(safe['CreditScore'], bins=30, alpha=0.6, label='Safe', color='green')
plt.hist(risky['CreditScore'], bins=30, alpha=0.6, label='Risky', color='red')

plt.title("Credit Score: Safe vs Risky Customers")
plt.xlabel("Credit Score")
plt.ylabel("Count")
plt.legend()
plt.show()
```

# Insights

Risky customers (red) slightly decrease as credit score increases

Safe and risky distributions remain mostly stable across all credit score ranges

Key insight: Credit score alone is not a strong predictor of default in this dataset, suggesting weak or simulated relationship between credit score and risk.

```python
plt.figure(figsize=(10,5))

plt.hist(safe['Income'], bins=30, alpha=0.6, label='Safe', color='green')
plt.hist(risky['Income'], bins=30, alpha=0.6, label='Risky', color='red')

plt.title("Income: Safe vs Risky Customers")
plt.xlabel("Income")
plt.ylabel("Count")
plt.legend()
plt.show()
```

#  Insights

Higher default risk is seen at lower income levels (below 40,000)
As income increases, risky cases decrease and become stable after around 60,000

Key insight: Income is a strong predictor of default at low levels, but its impact reduces once income reaches a middle range.

```python
import seaborn as sns

sns.scatterplot(data=df, x='Income', y='LoanAmount', hue='Default')
plt.title("Income vs Loan Amount (Safe vs Risky)")
plt.show()
```

#  Insights

The chart plots income against loan amount:

Blue (0): Safe customers who repaid loans
Orange (1): Risky customers who defaulted

Key insight: There is no clear separation between safe and risky customers using only income and loan amount together, even though both features show some risk patterns individually.

#  Final Conclusion

Individual features like income and loan amount show only weak patterns related to default risk.
    
Low income and high loan amounts slightly increase the chance of default, but the relationship is not strong or consistent.
    
When combining features, the data remains heavily mixed with no clear separation between safe and risky customers.
    
Simple rule-based decisions are not effective for this dataset due to overlapping patterns.
    
To accurately predict loan default, a Machine Learning classification model is required to capture hidden relationships in the data.
